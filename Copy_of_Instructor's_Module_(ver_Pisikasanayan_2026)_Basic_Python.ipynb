## Python has minimalist syntax.



As you may have noticed, Python doesn't implement the dreaded semicolon for every line of code, unlike with C++ and Java. In fact, Python follows this kind of philosophy in general, and uses very minimal syntax.

When nesting code, other low-level languages require the use of braces $\{\}$.

Python does not. Python nests your commands using **indentation.**

Observe the code bellow, there are no semicolons, and the looping element is not contained within braces, but rather by indentation.

````python
a_list = ["e", "l", "l", "o", "\n" , "w", "o", "r", "l", "d", "!"] # no semicolon!
print("H e l l o  w o r l d !")  # again, no semicolon!
for i in range(11):
  print(a_list[i]) # no braces, indented instead!
````

This works because Python is an **interpreted language**, that is, every single line of code is translated to assembly and then run, instead of the entire code being *compiled* and then run, such as in C++ or Fortran.

This is why you don't need semicolons to denote the end of a line of code, and why you don't need braces to denote lines of code included within nested code elements.

## Logical Operators.

There are three logical operators in Python:
* <code>and</code> returns True of both statements are true.
* <code>or</code> returns True if at least one statement is true.
* <code>not</code> inverses a Boolean. True into False and vice versa.

As you may or may not have noticed, logical operators are reliant on comparison operators to create conditional statements. (They're also reliant on identity and membership operators, which we will discuss briefly.)



````python
x = 2
y = -1

print(x == y and x != y) # Only one of these statements is true.
print(x > y and x != y) # Both statements are true.
print(x > y or x < y) # One of the statements is true.
print(not(x != y)) # The statement is true.
````

## Identity and Membership Operators.

Identity operators:
* <code>is</code> returns True if both objects are the same.
* <code>is not</code> returns True if both objects are not the same.

Membership operators:
* <code>in</code> returns True if value is in object.
* <code>not in</code> returns True if value is not in object.

````python
x = [1, 2, 3, 4, 5] # Oops, a list! This will be covered next topic.
y = 4 # Mention that hardcoding is possible, but not recommended.
z = 4
w = 1

print(y is z) # y is indeed z.
print(y is not w) # y is indeed not w.
print(y in x) # y is indeed in x.
print(w not in x) # w is actually in x.
````

## List Comprehension.

Other than the usual way of declaring Python variables, there's something called *list comprehension* which is a more compact way to create our lists.

````python
z_list = [i for i in range(0, 5)]
a_list = [j for j in range(-5, 0)]

print(z_list, a_list)

animal_list = ["dog", "cat", "parrot", "cheetah", "crocodile"]
c_list = [k for k in animal_list if "c" in k]

print(c_list)

frac_list = [i + 0.5 for i in range(1, 11)]
half_frac_list = [i for i in frac_list if i > 5.5]

print(frac_list, half_frac_list)
````

## List Slicing.

How are the elements in a list listed? That sounds like a dumb question, but we have to understand that Python doesn't see and count like we do. Also, it's going to be incredibly slow and inefficient if Python has to count from the top (or the first element of a list) just to find a certain element in a list.

So, how does Python do it?

Python uses indexing, which is basically assigning an integer number to a list element so that it knows exactly where it is on the list.

For example, in our list <code>animal_list</code>, the elements <code>"dog"</code>, <code>"cat"</code>, <code>"parrot"</code> are the first three elements. Thus, they have the indexes 0, 1, 2. (Python begins its indexing with 0 because it's essentially asking "How many positions to the right of the beginning is this element?" It follows that the first element is zero positions away from itself.)

Now, in Python when we define bounds it is almost always $[a, b)$; inclusive of the lower bound and exclusive of the upper bound. Hence, if we want to include the bound b, then we need to use the interval $[a, b+1)$.

List indices are the foundation of list slicing and manipulation. By specifying limits or bounds using indexing, we can slice lists:

````python
print(animal_list[0])
print(animal_list[0:5:2]) # There's a way to shorten this syntax, can you figure it out?
````

What we did in the code cell above is positive indexing, meaning we counted to the right from the beginning. If we did negative indexing, think of it as the index going to the left and then looping over to the end of the list, then going left from there.

In this case, Python is asking, "How many positions to the left of the beginning is this element?"

Going back to our example, <code>animal_list</code>, the -1 indexed element is actually the last element in the list. It should be apparent then how useful negative indexing is because now we can easily access the farther end of our list:

````python
print(animal_list[-1])
print(animal_list[:-1]) # There's a way to shorten this syntax, can you figure it out?
````

Lastly, when list slicing, we can also specify the step size of our indexing, this determines if we go element per element, every two elements, every three elements, etc.

````python
print(frac_list[:-1:2])
````

Now that we've learned conditional control through conditional statements, we can finally get into some iteration.

What is iteration?

Iteration is simply the repetition of an operation or process. When programming, we encounter many instances where we have to *iterate* or repeat a process, often many times. When I say many, I mean *many*.

Now, technically you can just copy and paste your instruction however many times you want. Sometimes, if the instruction is complex and repeats for around less than 5 iterations, that's faster then figuring out a loop. However, this is very rarely the case. No one wants to write ten thousand lines of code. And no one wants to read ten thousand lines of code. Thankfully those ten thousand lines can be simplified into less than ten with a simple loop.

````python
for i in range(100): # Print 100 "hello world!" messages.
  print("hello world!")
````

````python
# Print 100 "hello world!" messages.
i = 0
while i < 100:
  print("hello world!")
  i+=1
````

# Functions and Packages

## Functions Make Code Functional.

Lastly, for this basic tutorial, we'll discuss functions and packages in Python.

When we create a program (set of instructions to do something), it's often more convenient, efficient, and organized if we take that entire process and define it as a function. Why? Well, instead of having to write all those lines of code again (and yes, even when you copy paste, it's still tedious; not to mention it will make your code look horrendous), we can simply *call the function* in order to perform the same operation.

All of the processes and parts of Python we learned earlier can be included within functions! It should be apparent to you how useful this *function* of Python can be.

For example, say we want to slice many lists in half. Of course we can just write the code for all of them, but it's a much better practice to simply write the process once, define it as a function, and call it on all the lists we want to perform on.

````python
def list_slicer(l):
  x = int(len(l)/2)
  if len(l)%2 == 0:
    half_1 = l[:x]
    half_2 = l[x:]
    print(half_1, half_2)
  else:
    print("Your list is not even, bigger half 1 or half 2?")
    cond = int(input("half 1 = 1, half 2 = 2 \n"))
    if cond == 1:
      half_1 = l[:x+1]
      half_2 = l[x+1:]
      print(half_1, half_2)
    elif cond == 2:
      half_1 = l[:x]
      half_2 = l[x:]
      print(half_1, half_2)
    else:
      print("Input a valid number!")
````

````python
list_1 = [i for i in range(1, 11)]
list_2 = [i for i in range(1, 12)]
list_3 = [i for i in range(1, 21)]
list_4 = [i for i in range(10, 21)]
list_5 = [i for i in range(10, 22)]
````
